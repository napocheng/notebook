```python
# init libs and variables
import requests
from bs4 import BeautifulSoup as bs4

url = "http://www.cnblogs.com"
```

```python
# get the url content
res = requests.get(url)
soup = bs4(res.text, "lxml")

print(soup.prettify())
```

```python
# parser the html with css selector
links = soup.select(".current_nav")
for link in links:
    print(link)
```
