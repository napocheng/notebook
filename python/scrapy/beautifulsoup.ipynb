```python
# init libs and variables
import requests
from bs4 import BeautifulSoup as bs4

url = 'http://www.cnblogs.com'
```

```python
# get the url content
res = requests.get(url)
soup = bs4(res.text, "lxml")

print(soup.prettify())
```

```python
# parser the html with css selector
# CSS选择器 http://www.w3school.com.cn/cssref/css_selectors.asp
links = soup.select(".titlelnk")
for link in links:
    print(link.text)
```
